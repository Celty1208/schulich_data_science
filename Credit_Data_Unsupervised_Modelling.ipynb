#### <font color="#FF6B35">üç¨ ***Clustering Algorithms***</font>

```python
df.drop(['CUST_ID'],axis=1,inplace=True)
# Customer ID not necessary for pattern detection.
```

```python
# PCA for noise reduction.
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_std = scaler.fit_transform(df)

pca = PCA()
df_pca = pca.fit_transform(df_std)

df_pca = pd.DataFrame(df_pca, columns=[f'PC{i+1}' for i in range(len(df.columns))])

print('Explained variance ratio:', pca.explained_variance_ratio_)

cumsum_variance = np.cumsum(pca.explained_variance_ratio_)
print('Cumulative explained variance ratio:', cumsum_variance)

df_pca.head()
```

```python
pca = PCA(n_components = 10) 
df_pca = pca.fit_transform(df_std)

df_pca = pd.DataFrame(df_pca)

print('Explained variance ratio:', pca.explained_variance_ratio_)

cumsum_variance = np.cumsum(pca.explained_variance_ratio_)
print('Cumulative explained variance ratio:', cumsum_variance)

df_pca.head()
```

üéÉ With 9 components, 89% of the varability in data is explained.

```python
# Start clustering.
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
X = df_pca.to_numpy()
```

```python
# Elbow Method
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 11), inertia, color='#FF6B35')
plt.title('Elbow Method', fontsize=12, color='#E63946', fontweight='bold')
plt.xlabel('Number of clusters', fontsize=10, color='#FF6B35', fontweight='bold')
plt.ylabel('Inertia', fontsize=10, color='#FF6B35', fontweight='bold')
plt.show()
```

üéÉ K-Means Clustering has been applied, and the Elbow graph is displayed above. From the graph, choosing 4 clusters might be a good option to segment the customers in the dataset.

```python
# Applying K-means with 4 clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
```

```python
# Silhouette Score
sil_score = silhouette_score(X, labels)
print(f"Silhouette Score: {sil_score}")

plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centers[:, 0], centers[:, 1], c='red')
plt.title('K-means Clustering', fontsize=12, color='#E63946', fontweight='bold')
plt.show()
```

üéÉ The Silhouette Score of 0.2129 suggests that the clustering structure is weak and could be improved. While it's positive (indicating that, on average, samples are closer to their own cluster than to other clusters), it's not particularly close to 1.

üéÉ The K-means clustering plot shows that many scatter points are still dispersed but it is visually clustered into 4 groups.

```python
# Hierarchical Clustering
from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(X, method='ward')

# Plot dendrogram
plt.figure(figsize=(20, 5))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram',fontsize=12, color='#E63946', fontweight='bold')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

cluster_labels = fcluster(Z, t=50, criterion='distance')

# Compute Silhouette Score
hc_sil_score = silhouette_score(X, cluster_labels)
print(f"Hierarchical Clustering Silhouette Score: {hc_sil_score}")
```

üéÉ The Hierarachical Clustering shows a broad 4 clusters with color of yellow, green, red, and purple. It has a moderate Silhouette Score of 0.1676.

```python
# Affinity Propagation
from sklearn.cluster import AffinityPropagation

af = AffinityPropagation(random_state=0)
af_labels = af.fit_predict(X)

# Silhouette Score
af_sil_score = silhouette_score(X, af_labels)
print(f"Affinity Propagation Silhouette Score: {af_sil_score}")

plt.scatter(X[:, 0], X[:, 1], c=af_labels)
plt.title('Affinity Propagation Clustering',fontsize=12, color='#E63946', fontweight='bold')
plt.show()
```

üéÉ The Affinity Propagation shows very cluttered and unclear about the groups with a low Silhouette Score of 0.1398.

```python
# DBSCAN
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Silhouette Score
# Note: DBSCAN can result in noise data point with label -1
# Only include points that are part of a cluster
if len(set(dbscan_labels)) > 1:
    dbscan_sil_score = silhouette_score(X, dbscan_labels)
    print(f"DBSCAN Silhouette Score: {dbscan_sil_score}")

# Plotting
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels)
plt.title('DBSCAN Clustering',fontsize=12, color='#E63946', fontweight='bold')
plt.show()
```

üéÉ The DBCAN did not perform very well with the graph cluttered and negative Silhouette Score of -0.4982.

```python
df_score = pd.DataFrame({"K-means Silhouette Score": [sil_score],
"Hierarchical Clustering Silhouette Score": [hc_sil_score],
"Affinity Propagation Silhouette Score": [af_sil_score],
"DBSCAN Silhouette Score": [dbscan_sil_score]})
```
