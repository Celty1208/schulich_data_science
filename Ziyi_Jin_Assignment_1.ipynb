```python
#Define a method to fill the missing unit price value 
#Find the mean of Unit Price for every product category from 1 to 50
#fill in the N/A Unit price using the mean for every product.

def fill_missing_prices(df, product):
    product_condition = df['Product'] == product
    df.loc[product_condition, 'Price'] = df.loc[product_condition, 'Price'].fillna(df[df['Product'] == product]['Price'].mean())
```

```python
#Fill all the unit prices using the extracted method.
for products in df['Product']:
    fill_missing_prices(df, products)
```

```python
#Find the top3 products in each region.
grouped_data=df.groupby(['Product','Region'])
sum_table=grouped_data['Units Sold'].sum().reset_index()
sum_grouped=sum_table.groupby('Region')
```

```python
list=[]
regions=['Region_01','Region_02','Region_03', 'Region_04', 'Region_05', 'Region_06', 'Region_07', 'Region_08', 'Region_09', 'Region_10']

for region in regions:
    region_data=sum_grouped.get_group(region)
    top3_products=region_data.head(3)
    list.extend(top3_products.to_dict('records'))
top3_table=pd.DataFrame(list)
top3_table
```

```python
plt.figure(figsize=(7,5))
top3_table['Product_Category']=top3_table['Product']+' '+'('+top3_table['Region']+')'
sns.set(style="whitegrid")
colors=sns.color_palette("pastel",len(top3_table["Region"].unique()))
sns.barplot(x='Product_Category',y='Units Sold',hue='Region',width=1,palette=colors,data=top3_table,ci=None,dodge=0,alpha=0.8)
plt.title('Top3 Products by Region',fontsize=14)
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.xticks(rotation=90)
plt.legend(bbox_to_anchor=(1,1))
plt.show()
```

The chart depicts the top-selling products in different regions. Based on the chart, it is evident that Product 1 and 10 are highly popular among all of the regions.

Hot-selling products can often be closely linked to the local region, such as region-specific essentials. Therefore, companies can push different hot-selling products on social media based on the region to increase their exposure and achieve an increase in sales.
